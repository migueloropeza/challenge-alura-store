### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresoTotalTienda1 = sum(tienda["Precio"])
print(f'El ingreso total de la tienda 1 es: {ingresoTotalTienda1}')
ingresoTotalTienda2 = sum(tienda2["Precio"])
print(f'El ingreso total de la tienda 2 es: {ingresoTotalTienda2}')
ingresoTotalTienda3 = sum(tienda3["Precio"])
print(f'El ingreso total de la tienda 3 es: {ingresoTotalTienda3}')
ingresoTotalTienda4 = sum(tienda4["Precio"])
print(f'El ingreso total de la tienda 4 es: {ingresoTotalTienda4}')
```

```python
tiendaMayorIngreso = max(sum(tienda["Precio"]),sum(tienda2["Precio"]),sum(tienda3["Precio"]),sum(tienda4["Precio"]))
if tiendaMayorIngreso == ingresoTotalTienda1:
  print(f'La tienda 1 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
elif tiendaMayorIngreso == ingresoTotalTienda2:
  print(f'La tienda 2 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
elif tiendaMayorIngreso == ingresoTotalTienda3:
  print(f'La tienda 3 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
elif tiendaMayorIngreso == ingresoTotalTienda4:
  print(f'La tienda 4 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
```

```python
plt.bar(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], [ingresoTotalTienda1, ingresoTotalTienda2, ingresoTotalTienda3, ingresoTotalTienda4], color=["red", "blue", "green", "yellow"])
plt.title("Ingresos por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

for i, v in enumerate([ingresoTotalTienda1, ingresoTotalTienda2, ingresoTotalTienda3, ingresoTotalTienda4]):
    plt.text(i, v + 5000000, str(v), ha='center', va='bottom', rotation=0)

plt.show()
```

# 2. Ventas por categoría

```python
agrupacionTienda1 = tienda.groupby("Categoría del Producto")
ventasCategoriaTienda1 = agrupacionTienda1["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 1 son: {ventasCategoriaTienda1.sort_values(ascending=False)}')
agrupacionTienda2 = tienda2.groupby("Categoría del Producto")
ventasCategoriaTienda2 = agrupacionTienda2["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 2 son: {ventasCategoriaTienda2.sort_values(ascending=False)}')
agrupacionTienda3 = tienda3.groupby("Categoría del Producto")
ventasCategoriaTienda3 = agrupacionTienda3["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 3 son: {ventasCategoriaTienda3.sort_values(ascending=False)}')
agrupacionTienda4 = tienda4.groupby("Categoría del Producto")
ventasCategoriaTienda4 = agrupacionTienda4["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 4 son: {ventasCategoriaTienda4.sort_values(ascending=False)}')
```

# 3. Calificación promedio de la tienda


```python
calificacionPromedioTienda1 = tienda['Calificación'].mean()
print(f'La calificación promedio de la tienda 1 es: {calificacionPromedioTienda1}')
calificacionPromedioTienda2 = tienda2['Calificación'].mean()
print(f'La calificación promedio de la tienda 2 es: {calificacionPromedioTienda2}')
calificacionPromedioTienda3 = tienda3['Calificación'].mean()
print(f'La calificación promedio de la tienda 3 es: {calificacionPromedioTienda3}')
calificacionPromedioTienda4 = tienda4['Calificación'].mean()
print(f'La calificación promedio de la tienda 4 es: {calificacionPromedioTienda4}')
```

```python
plt.bar(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], [calificacionPromedioTienda1,calificacionPromedioTienda2, calificacionPromedioTienda3, calificacionPromedioTienda4], color=["red", "blue", "green", "yellow"])
plt.title("Calificaciones promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificaciones")
for i, v in enumerate([calificacionPromedioTienda1.round(2), calificacionPromedioTienda2.round(2), calificacionPromedioTienda3.round(2), calificacionPromedioTienda4.round(2)]):
    plt.text(i, v , str(v), ha='center', va='bottom', rotation=0)
plt.show()
```

# 4. Productos más y menos vendidos

```python
productosVendidosTienda1 = tienda.groupby("Producto")
ventasProductosTienda1 = productosVendidosTienda1["Producto"].count()
print(f'Las ventas por producto de la tienda 1 son: {ventasProductosTienda1.sort_values(ascending=False)}')
productosVendidosTienda2 = tienda2.groupby("Producto")
ventasProductosTienda2 = productosVendidosTienda2["Producto"].count()
print(f'Las ventas por producto de la tienda 2 son: {ventasProductosTienda2.sort_values(ascending=False)}')
productosVendidosTienda3 = tienda3.groupby("Producto")
ventasProductosTienda3 = productosVendidosTienda3["Producto"].count()
print(f'Las ventas por producto de la tienda 3 son: {ventasProductosTienda3.sort_values(ascending=False)}')
productosVendidosTienda4 = tienda4.groupby("Producto")
ventasProductosTienda4 = productosVendidosTienda4["Producto"].count()
print(f'Las ventas por producto de la tienda 4 son: {ventasProductosTienda4.sort_values(ascending=False)}')
```

```python
plt.plot(ventasProductosTienda1.sort_values(ascending=False).head(), marker='o', linestyle='', color="red")
plt.plot(ventasProductosTienda2.sort_values(ascending=False).head(), marker='o', linestyle='', color="blue")
plt.plot(ventasProductosTienda3.sort_values(ascending=False).head(), marker='o', linestyle='', color="green")
plt.plot(ventasProductosTienda4.sort_values(ascending=False).head(), marker='o', linestyle='', color="yellow")
plt.title("Productos mas vendidos por Tienda")
plt.xlabel("Productos")
plt.ylabel("Cantidad")
plt.xticks(rotation=90)
plt.grid()
plt.legend(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
plt.show()
plt.plot(ventasProductosTienda1.sort_values(ascending=True).head(), marker='o', linestyle='', color="red")
plt.plot(ventasProductosTienda2.sort_values(ascending=True).head(), marker='o', linestyle='', color="blue")
plt.plot(ventasProductosTienda3.sort_values(ascending=True).head(), marker='o', linestyle='', color="green")
plt.plot(ventasProductosTienda4.sort_values(ascending=True).head(), marker='o', linestyle='', color="yellow")
plt.title("Productos menos vendidos por Tienda")
plt.xlabel("Productos")
plt.ylabel("Cantidad")
plt.xticks(rotation=90)
plt.grid()
plt.legend(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
plt.show()
```

###Introducción

El propósito de este análisis es determinar cuál tienda es la más adecuada para que el Sr. Juan enfoque sus esfuerzos de venta, basándose en un conjunto integral de factores clave. Estos incluyen los ingresos totales, las categorías de productos con mayor y menor demanda, las calificaciones promedio de los clientes, los productos más y menos vendidos, y el coste de envío promedio. La finalidad es identificar una tienda que ofrezca un balance óptimo entre rentabilidad, satisfacción del cliente y eficiencia operativa para maximizar el éxito comercial.

##🧠 Análisis Crítico Combinado

Tienda 1 lidera en ingresos y variedad de productos, lo cual evidencia una estrategia comercial fuerte. Sin embargo, presenta la calificación más baja de satisfacción del cliente, combinada con el costo de envío más alto. Esto sugiere una baja percepción de valor por parte del cliente, lo que podría traducirse en fugas de ingresos a futuro si no se corrigen estos problemas.

Tienda 3 tiene una satisfacción del cliente sobresaliente, lo cual es un activo importante para fidelización y crecimiento sostenible. Aunque sus ingresos no son los más altos, su desempeño es sólido y con buen potencial.

Tienda 4, si bien tiene el costo de envío más bajo y una buena cobertura, es la que presenta peor desempeño general en ingresos, rotación de productos y competitividad. Sus productos no destacan frente a los de otras tiendas, y su demanda parece limitada por factores externos como la ubicación o la competencia local.

Tienda 2 muestra indicadores estables y constantes, sin puntos críticos ni destacados, lo cual la hace menos riesgosa a corto plazo y con espacio para mejoras.
