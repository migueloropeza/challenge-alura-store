### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresoTotalTienda1 = sum(tienda["Precio"])
print(f'El ingreso total de la tienda 1 es: {ingresoTotalTienda1}')
ingresoTotalTienda2 = sum(tienda2["Precio"])
print(f'El ingreso total de la tienda 2 es: {ingresoTotalTienda2}')
ingresoTotalTienda3 = sum(tienda3["Precio"])
print(f'El ingreso total de la tienda 3 es: {ingresoTotalTienda3}')
ingresoTotalTienda4 = sum(tienda4["Precio"])
print(f'El ingreso total de la tienda 4 es: {ingresoTotalTienda4}')
```

```python
tiendaMayorIngreso = max(sum(tienda["Precio"]),sum(tienda2["Precio"]),sum(tienda3["Precio"]),sum(tienda4["Precio"]))
if tiendaMayorIngreso == ingresoTotalTienda1:
  print(f'La tienda 1 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
elif tiendaMayorIngreso == ingresoTotalTienda2:
  print(f'La tienda 2 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
elif tiendaMayorIngreso == ingresoTotalTienda3:
  print(f'La tienda 3 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
elif tiendaMayorIngreso == ingresoTotalTienda4:
  print(f'La tienda 4 cuenta con el mayor ingreso con un monto de: {tiendaMayorIngreso}')
```

# 2. Ventas por categoría

```python
agrupacionTienda1 = tienda.groupby("Categoría del Producto")
ventasCategoriaTienda1 = agrupacionTienda1["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 1 son: {ventasCategoriaTienda1}')
agrupacionTienda2 = tienda2.groupby("Categoría del Producto")
ventasCategoriaTienda2 = agrupacionTienda2["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 2 son: {ventasCategoriaTienda2}')
agrupacionTienda3 = tienda3.groupby("Categoría del Producto")
ventasCategoriaTienda3 = agrupacionTienda3["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 3 son: {ventasCategoriaTienda3}')
agrupacionTienda4 = tienda4.groupby("Categoría del Producto")
ventasCategoriaTienda4 = agrupacionTienda4["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 4 son: {ventasCategoriaTienda4}')
```

```python
agrupacionTienda1 = tienda.groupby("Categoría del Producto")
ventasCategoriaTienda1 = agrupacionTienda1["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 1 son: {ventasCategoriaTienda1.sort_values(ascending=False)}')
agrupacionTienda2 = tienda2.groupby("Categoría del Producto")
ventasCategoriaTienda2 = agrupacionTienda2["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 2 son: {ventasCategoriaTienda2.sort_values(ascending=False)}')
agrupacionTienda3 = tienda3.groupby("Categoría del Producto")
ventasCategoriaTienda3 = agrupacionTienda3["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 3 son: {ventasCategoriaTienda3.sort_values(ascending=False)}')
agrupacionTienda4 = tienda4.groupby("Categoría del Producto")
ventasCategoriaTienda4 = agrupacionTienda4["Categoría del Producto"].count()
print(f'Las ventas por categoria de la tienda 4 son: {ventasCategoriaTienda4.sort_values(ascending=False)}')
```

# 3. Calificación promedio de la tienda


```python
calificacionPromedioTienda1 = tienda['Calificación'].mean()
print(f'La calificación promedio de la tienda 1 es: {calificacionPromedioTienda1}')
calificacionPromedioTienda2 = tienda2['Calificación'].mean()
print(f'La calificación promedio de la tienda 2 es: {calificacionPromedioTienda2}')
calificacionPromedioTienda3 = tienda3['Calificación'].mean()
print(f'La calificación promedio de la tienda 3 es: {calificacionPromedioTienda3}')
calificacionPromedioTienda4 = tienda4['Calificación'].mean()
print(f'La calificación promedio de la tienda 4 es: {calificacionPromedioTienda4}')
```

# 4. Productos más y menos vendidos

```python
productosVendidosTienda1 = tienda.groupby("Producto")
ventasProductosTienda1 = productosVendidosTienda1["Producto"].count()
print(f'Las ventas por producto de la tienda 1 son: {ventasProductosTienda1.sort_values(ascending=False)}')
productosVendidosTienda2 = tienda2.groupby("Producto")
ventasProductosTienda2 = productosVendidosTienda2["Producto"].count()
print(f'Las ventas por producto de la tienda 2 son: {ventasProductosTienda2.sort_values(ascending=False)}')
productosVendidosTienda3 = tienda3.groupby("Producto")
ventasProductosTienda3 = productosVendidosTienda3["Producto"].count()
print(f'Las ventas por producto de la tienda 3 son: {ventasProductosTienda3.sort_values(ascending=False)}')
productosVendidosTienda4 = tienda4.groupby("Producto")
ventasProductosTienda4 = productosVendidosTienda4["Producto"].count()
print(f'Las ventas por producto de la tienda 4 son: {ventasProductosTienda4.sort_values(ascending=False)}')
```
